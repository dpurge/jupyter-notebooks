# OCR Example

```python
from ipython_helpers import display_text, display_image
from ocr_helpers import image_from_url, image_from_pdf, ocr_image
```

```python
arb = image_from_url('http://.../page-182.png')
```

```python
display_image(arb)
```

```python
text = ocr_image(arb, "ara")
```

```python
display_text(text, style='direction: rtl; text-align: right; font-size: 22px;')
```

```python
ell = image_from_pdf('../data/greek1.pdf', 152)
```

```python
display_image(ell)
```
